```python
# groupby: spliting a dataframe on values of categorical variables
groupby_gender=data.groupby("Gender")
for gender, value in groupby_gender['VIQ']:
    print(gender, value.mean())
```

```python
# groupby_gender is a powerful tool that expose many operations on the resulting group of dataframe
groupby_gender.mean()
```

```python
groupby_gender.median()
groupby_gender.sum()
```

```python
# What is the mean value for VIQ for the full population?
data['VIQ'].mean()
```

```python
# How many males/females were included in this study?
groupby_gender.count()
```

```python
# What is the average value of MRI counts expressed in log units, for males and females?
for gender, value in groupby_gender['MRI_Count']:
    print(gender, value.mean())
```
